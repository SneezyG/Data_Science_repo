from the look of things stock and colour columns are completely null, but i need them, 
let me do something about that.

```python
# let me generate some random color;
np.random.seed(0)
color = ["red", "blue", "light blue", 
          "green", "orange", "pink",
         "purple", "brown", "black", 
         "white", "grey"]
color = pd.Series(color)
colors = color[np.random.randint(10, 
                   size=(10002))]
print("color: ", colors.values)
print("color count: ", colors.count())
```

```python
# let generate some random stock number;
np.random.seed(0)
stock = pd.Series(np.random.randint(5, 40,
                          size=(10002)))
print("stock: ", stock.values)
print("stock count: ", stock.size)
```

And there we have it, my color and stock fill is ready

```python
"""
let fill in nan in color and stock column;
in product;
"""
product.Stock = stock
product.Color = colors.values
```

```python
# let me have a glimpse of the new dataset;
product.head()
```

```python
product.info()
```
