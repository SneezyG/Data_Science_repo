```python
import numpy as np;
import pandas as pd;
```

```python
# load the data with pandas;
path = "/storage/emulated/0/Download/Olympics.csv"
olympic = pd.read_csv(path)
```

```python
# let have a glimpse of the dataset
olympic.head()
```

```python
"""
let get list of all country to ever
participate in the Olympic, the country
column is label NOC;
"""
olympic["NOC"]
```

```python
"""
let's create a pandas series of continent
from the dataset with the number of times 
they have participate in olympic through
the countries in them;
"""
continents = np.array(olympic["Continent"])
count = {}
for x in continents:
    try:
        count[x] += 1
    except:
        count[x] = 1
continent = pd.Series(count)
continent.index.name = "Continent"
continent
```

```python
"""
the computation like above can easily be done 
using pandas with pivot table;
"""
continent = olympic.pivot_table(
               index="Continent",
               aggfunc= {"NOC":"count"})
continent
```

```python
# let's take care of the missing data;
# lets fill in misisng data with 0;
olympic = olympic.fillna(0)
olympic.head()
```

```python
# we can get a summarised info about olympic dataset;
olympic.info()
```

```python
# a summarized discription of the data;
# it's quite discriptive
olympic.describe()
```

```python
"""
lets create a multi index dataset out of
the olympic dataset;
"""
#select record that have europe as their continent
euro = olympic[olympic.Continent ==
               "Europe"]

#and record with more than 20 total medals
euro = euro[euro["Total Medals"] > 15]
euro
```

```python
# let's select the column we are interested in
array = np.array(euro[[
            "Gold",
            "Silver",
            "Bronze",
        ]])
array
```

```python
# this create columns name for the multindex
columns = pd.MultiIndex.from_product([['Medals'],
            ["gold", "silver", "bronze",
              ]],
          names=['subject', 'type'])
```

```python
# this create index name for the multindex
index = np.array(euro.NOC)
```

```python
# let's get it, this finally does the job
medal = pd.DataFrame(array, index= index,
               columns=columns)
medal.index.name = "country"
medal
```

```python
# indexing and slicing in multi index;
medal.loc["Sweden":"Austria"]
```

```python
# slicing with the column;
medal["Medals", "silver"]
```

```python
# data aggregation with multi index;
medal.sum(axis=0)
```

```python
"""
concate and append in pandas;
let's try this method by creating
another multiindex dataset;
"""
array = np.array(euro[[
            "Men Athletes",
            "Women Athletes"
        ]])

index = np.array(euro.NOC)

column = pd.MultiIndex.from_product([["Athletes"],
             ["men", "women"]],
             names=["subject", "type"])

athlete = pd.DataFrame(array,
                 index=index, 
                 columns = column)
athlete
```

```python
# let's append the two multindex aray;
data = pd.concat([medal,athlete], axis=1)
data
```

```python
# grouping and aggregate with pandas;
# life become simpler;
# let's group our olympic data on gold medals;
olympic3 = olympic.groupby("Continent")[["Gold"]].sum()  
olympic3
```

```python
"""
working with pandas eval and query;
they make computation on large data a walk
in a park;
"""
x = pd.Series(np.random.randint(7,
                        size=(4)))
y = pd.Series(np.random.randint(4, 9,
                        size=(4)))
%timeit pd.eval(x+y)
%timeit x + y
```

```python
print(y, y.sum())
print(x, x.sum())
```

```python
# eval can also be use for checking conditions;
pd.eval("(x.sum() < y.sum()) & (y.min()>x.min())")
```

```python
"""
let's move to pandas query method;
write long query the easy way
it let you execute large dataset query with speed advantages.        
let get our euro;
let's get the olympic dataset;
"""
olympic
```

```python
"""
let's query the dataset for Europe countries
with 8 golds and 5 silvers
"""
olympic.query("(Gold > 8 and Silver > 5) & (Continent =='Europe')")
```
